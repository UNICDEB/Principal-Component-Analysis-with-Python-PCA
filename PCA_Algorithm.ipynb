```python
import numpy as np
import pandas as pd

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

1) Import MNIST dataset

```python
# Importing MNSIT Data

from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
mnist.target = mnist.target.astype(np.int64)
from sklearn.model_selection import train_test_split

X = mnist["data"]
y = mnist["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Plotting a random image

some_digit = X[46]
some_digit_image = some_digit.reshape(28, 28)
plt.imshow(some_digit_image, cmap="binary")
plt.axis("off")
plt.show()
```

```python
# Finding number of components

pca = PCA().fit(X_train)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

2) Apply PCA retain principal components with variability 80%,83%, 85%,90%, 92% , 95%

```python
#  Apply PCA retain principal components with variability 80%

pca = PCA(n_components=0.80)
X_reduced01 = pca.fit_transform(X_train)
X_test01=pca.transform(X_test)
pca.n_components_
```

```python
#  Apply PCA retain principal components with variability 83%

pca = PCA(n_components=0.83)
X_reduced02 = pca.fit_transform(X_train)
X_test02=pca.transform(X_test)
pca.n_components_
```

```python
#  Apply PCA retain principal components with variability 85%

pca = PCA(n_components=0.85)
X_reduced03 = pca.fit_transform(X_train)
X_test03=pca.transform(X_test)
pca.n_components_
```

```python
#  Apply PCA retain principal components with variability 90%
pca = PCA(n_components=0.90)
X_reduced04 = pca.fit_transform(X_train)
X_test04=pca.transform(X_test)
pca.n_components_
```

```python
#  Apply PCA retain principal components with variability 92%

pca = PCA(n_components=0.92)
X_reduced05 = pca.fit_transform(X_train)
X_test05=pca.transform(X_test)
pca.n_components_
```

```python
#  Apply PCA retain principal components with variability 95%

pca = PCA(n_components=0.95)
X_reduced06 = pca.fit_transform(X_train)
X_test06=pca.transform(X_test)
pca.n_components_
```

3) Apply classification algorithm (Decision Tree, kNN) corresponding all the above six scenarios and report the classification accuracy

                                  Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Decision Tree with 80% variability

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_reduced01, y_train)
y_pred_tree = tree_clf.predict(X_test01)

# ACCURACY SCORE

print(accuracy_score(y_test, y_pred_tree))


```

```python
# Decision Tree with 83% variability

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_reduced02, y_train)
y_pred_tree = tree_clf.predict(X_test02)

# ACCURACY SCORE

print(accuracy_score(y_test, y_pred_tree))
```

```python
# Decision Tree with 85% variability

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_reduced03, y_train)
y_pred_tree = tree_clf.predict(X_test03)

# ACCURACY SCORE

print(accuracy_score(y_test, y_pred_tree))
```

```python
# Decision Tree with 90% variability

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_reduced04, y_train)
y_pred_tree = tree_clf.predict(X_test04)

# ACCURACY SCORE

print(accuracy_score(y_test, y_pred_tree))
```

```python
# Decision Tree with 92% variability

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_reduced05, y_train)
y_pred_tree = tree_clf.predict(X_test05)

# ACCURACY SCORE

print(accuracy_score(y_test, y_pred_tree))
```

```python
# Decision Tree with 95% variability

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_reduced06, y_train)
y_pred_tree = tree_clf.predict(X_test06)

# ACCURACY SCORE

print(accuracy_score(y_test, y_pred_tree))
```
